```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

model = Sequential()
model.add(Dense(25, input_dim=x_test.shape[1], activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, 
                        patience=5, verbose=1, mode='auto', 
                        restore_best_weights=True)

```

```python
model.fit(x_train, y_train, verbose=2, epochs=1000)
```

```python
from sklearn import metrics

# Predict
pred = model.predict(x_test)
print(pred)
# Measure MSE error.  
# score = metrics.mean_squared_error(pred,y_test)
# print("Final score (MSE): {}".format(score))
```

```python
# for a14

import pandas as pd

df_og = pd.read_csv(dataset, na_values=['NA','?'])

keep_cols = ['a14','s3','a8','a9','a10','a11','a12','a13','a15']
for i in df_og.columns:
    if i not in keep_cols:
        df_og.drop(i, inplace=True, axis=1)
```

```python
df_og = pd.get_dummies(df_og)
```

```python
df_test = df_og[df_og['a14'].isna()].copy()
```

```python
df_train = df_og.drop(df_og[df_og['a14'].isna()].index)
```

```python
x_train, y_train = df_train.drop('a14',axis=1), df_train['a14'].values
x_test, y_test = df_test.drop('a14',axis=1), df_test['a14'].values
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

model = Sequential()
model.add(Dense(50, input_dim=x_test.shape[1], activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, 
                        patience=5, verbose=1, mode='auto', 
                        restore_best_weights=True)

```

```python
model.fit(x_train, y_train, verbose=2, epochs=1000)
```

```python
# Predict
pred = model.predict(x_test)
print(pred)
```
