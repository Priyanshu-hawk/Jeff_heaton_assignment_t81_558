```python
# assignment_url = "https://github.com/jeffheaton/t81_558_deep_learning/blob/master/assignments/assignment_yourname_class5.ipynb"

hp_train = "https://data.heatonresearch.com/data/t81-558/datasets/houses_train.csv" # House price train
hp_test = "https://data.heatonresearch.com/data/t81-558/datasets/houses_test.csv" # House price test
hp_zip_median = "https://data.heatonresearch.com/data/t81-558/datasets/zips.csv" # Median Income by Zipcode
```

```python
import pandas as pd

df_train = pd.read_csv(hp_train, na_values=['?','NA'])
df_test = pd.read_csv(hp_test, na_values=['?','NA'])
df_med_inc = pd.read_csv(hp_zip_median, na_values=['?','NA'])
```

```python
#droping 'id'
df_train.drop('id', axis=1, inplace=True)
df_test.drop('id', axis=1, inplace=True)
```

```python
# key value pair for zip and income
median_income = {}
for i in range(len(df_med_inc)):
    median_income[df_med_inc.iloc[i]['zip']] = df_med_inc.iloc[i]['median_income']
```

```python
# including median_income using med_inc dataframe

df_train['median_income'] = df_train['zip'].map(median_income)
```

```python
# for train

y = df_train['price'].values
x = df_train.drop('price', axis=1).copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train
```

```python
## model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Flatten
import tensorflow as tf

model = Sequential()
# model.add(Flatten(input_shape=[x_train.shape[1]]))
model.add(Dense(90, activation='relu', input_dim=x_train.shape[1]))
model.add(Dense(50, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', metrics=[tf.keras.metrics.RootMeanSquaredError()], optimizer='adam')
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), verbose=2, epochs=1000)
```

```python
# RMSE is root_mean_squared_error(2084.9004) < 10000 which means model is ok. 
```
