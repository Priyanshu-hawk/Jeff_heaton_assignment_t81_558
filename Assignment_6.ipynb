```python
def sqr_frame_pading(img):
    
    row = img.shape[0]
    col = img.shape[1]
    r = max(row,col)
    c = r
    #print(row, col)

    avg_img = np.average(img, axis=(0,1))
    
    canvas = np.zeros((r,c,3), dtype=np.uint8)
    canvas[:,:] = avg_img
    
    sp = (r - col) // 2
    ep = (c - row) // 2
    
    #print(ep,sp)
    
    canvas[ep:row+ep,sp:col+sp] = img #placing image at center
    return canvas
```

```python
import requests
import numpy as np
from matplotlib.pyplot import imshow
from io import BytesIO
from PIL import Image

r = requests.get(img_url)
img = Image.open(BytesIO(r.content))

img_array = np.asarray(img)
img2_arr = sqr_frame_pading(img_array)
img2 = Image.fromarray(img2_arr, 'RGB')
img2
```
